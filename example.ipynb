### load data

```python
from analysetext import TextAnalysisPipeline
import pandas as pd

df=pd.read_csv('data/sampledata.csv',index_col='sentID')
```

### instantiate pipeline

pipeline returns:
1. term by cluster index df
2. terms by cluster index df
3. cluster plot

```python
pipeline=TextAnalysisPipeline()

termcluster,termscluster,p=pipeline.run(documents=df.sentence.values,indices=df.index)
```
